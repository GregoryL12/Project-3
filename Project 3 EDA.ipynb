# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
year0 = pd.read_csv(r"C:\Users\gleal\Desktop\Jnb Repo\Project-3\Data\final_tmdbdata2000.csv.gz")
year0.head()
```

```python
year1 = pd.read_csv(r"C:\Users\gleal\Desktop\Jnb Repo\Project-3\Data\final_tmdbdata2001.csv.gz")
year1.head()
```

```python
df_years = pd.concat([year0, year1],
                      ignore_index=True)
display(df_years.head(3), df_years.tail(3))
```

 ## 1. How many movies had at least some valid financial information (values > 0 for budget OR revenue)?

```python
df_movie_finance = df_years[(df_years['budget'] > 0) | (df_years['revenue'] > 0)]
df_movie_finance.head()
```

```python
print(f'The number of movies that had at least some valid financial information is {len(df_movie_finance)}')
```

## 2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
df_years["certification"].value_counts()
```

```python
sns.countplot(x=df_years["certification"]);
```

## 3. What is the average revenue per certification category?

```python
df_movie_revenue = df_years[df_years['revenue'] > 0]
```

```python
pd.options.display.float_format = '{:,.2f}'.format

rev_cert = df_movie_revenue.groupby('certification')['revenue'].mean()
rev_cert
```

## 4.What is the average budget per certification category?

```python
df_movie_budget = df_years[df_years['budget'] > 0]
```

```python
bud_cert = df_movie_budget.groupby('certification')['budget'].mean()
bud_cert
```
