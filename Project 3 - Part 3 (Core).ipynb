## I) Normalizing Genres

```python
df = pd.read_csv('Data/title_basics.csv.gz')
df.head()
```

```python
## create a col with a list of genres
df['genres_split'] = df['genres'].str.split(',')
df.head()
```

```python
df = df.explode('genres_split')
df
```

```python
unique_genres = sorted(df['genres_split'].unique())
```

```python
title_genres = df[['tconst', 'genres_split']].copy()
title_genres.head()
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
## make new integer genre_id and drop string genres
df['genre_id'] = df['genres_split'].replace(genre_map)
df = df.drop(columns='genres_split')
df.head()
```

```python
title_genres = df[['tconst', 'genre_id']].copy()
title_genres.head()
```

### Title Basics

```python
from sqlalchemy.types import *
## Calculate max string lengths for object columns
key_len = basics['tconst'].fillna('').map(len).max()
title_len = basics['primaryTitle'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "tconst": String(key_len+1), 
    "primaryTitle": Text(title_len+1),
    'startYear':Float(),
    'endYear':Float(),
    'runtimeMinutes':Integer()}
```

```python
# Save to sql with dtype and index=False
basics.to_sql('title_basics',engine,dtype=df_schema,if_exists='replace',index=False)
```

```python
engine.execute('ALTER TABLE title_basics ADD PRIMARY KEY (`tconst`);')
```

### Title Ratings

```python
## Calculate max string lengths for object columns
key_len_r = ratings['tconst'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema1 = {
    "tconst": String(key_len_r+1), 
    'averageRating':Float(),
    'numVotes':Integer()}
```

```python
# Save to sql with dtype and index=False
ratings.to_sql('title_ratings',engine,dtype=df_schema1,if_exists='replace',index=False)
```

```python
engine.execute('ALTER TABLE title_ratings ADD PRIMARY KEY (`tconst`);')
```

### TMDB Data

```python
## Calculate max string lengths for object columns
key_len_t = results['imdb_id'].fillna('').map(len).max()
cert_len = results['certification'].fillna('').map(len).max()

## Create a schema dictonary using Sqlalchemy datatype objects
df_schema2 = {
    "imdb_id": String(key_len_t+1),
    'budget':Float(),
    'revenue':Float(),
    "certification": Text(cert_len+1),
    }
```

```python
# Save to sql with dtype and index=False
results.to_sql('tmdb_data',engine,dtype=df_schema2,if_exists='replace',index=False)
```

```python
engine.execute('ALTER TABLE tmdb_data ADD PRIMARY KEY (`imdb_id`);')
```
